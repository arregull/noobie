```python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
def keihls():
    res = requests.get('https://www.kiehls.com/mens/view-all-men/')
    soup = BeautifulSoup(res.text,'lxml')
    items = soup.findAll('div',{'class':'product_tile b-product_tile'})
    
    product_name = [item.a.img['title'] for item in items]
    price = [item.p.text.split('\n')[5] for item in items]
    
    descriptions = soup.findAll('div',{'class':'product_description'})
    desc = [description.text.split('\n')[1] for description in descriptions]
    
    keihls_list = pd.DataFrame(list(zip(product_name,price,desc)),columns=['Product Name','Price','Description'],)
    return keihls_list.sort_values(by='Price',ascending=False).set_index('Product Name')
```

```python
keihls()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pic = pd.read_csv('Ariel_Patrick_Star.csv', index_col=0).fillna('')
```

```python
x = [pic[pic[str(i)]!=""][str(i)].apply(lambda x:x.split(',')[0]) for i in range(len(pic))]
```

```python
y = [pic[pic[str(i)]!=""][str(i)].apply(lambda x:x.split(',')[1]) for i in range(len(pic))]
```
